```python
def male_female_child(passenger):
    # Take the Age and Sex
    age,sex = passenger
    # Compare the age, otherwise leave the sex
    if age < 16:
        return 'child'
    else:
        return sex
    
    
```

```python
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)

titanic_df[0:10]
```

```python
sns.catplot(x= "Pclass", kind="count", data=titanic_df,hue='person')
```

```python
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
# Set empty list
levels = []

# Loop to grab first letter
for level in deck:
    levels.append(level[0])    

# Reset DataFrame and use factor plot
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
sns.catplot(x= "Cabin", kind="count",order=["A", "B","C","D","E","F","G","T"], data=cabin_df, palette="winter_d")
```

```python
# Redefine cabin_df as everything but where the row was equal to 'T'
cabin_df = cabin_df[cabin_df.Cabin != 'T']
#Replot
sns.catplot(x= "Cabin", kind="count", order=["A", "B","C","D","E","F","G"],data=cabin_df, palette="summer")
```

```python
titanic_df['Alone'] =  titanic_df.Parch + titanic_df.SibSp
titanic_df['Alone']
```

```python
# Look for >0 or ==0 to set alone status
titanic_df['Alone'].loc[titanic_df['Alone'] >0] = 'With Family'
titanic_df['Alone'].loc[titanic_df['Alone'] == 0] = 'Alone'
```
